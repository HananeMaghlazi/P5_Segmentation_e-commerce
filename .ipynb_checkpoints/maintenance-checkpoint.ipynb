```python
start_date = min(df_olist.order_purchase_timestamp)
end_date = max(df_olist.order_purchase_timestamp)

# Number of Months Between Two Dates
time_delta = int((end_date - start_date) / np.timedelta64(1, "M"))
print("La période totale est sur", time_delta, "mois")
```

```python
# Aggéger le dataframe avec les variables selectionnées
df_olist_2 = df_olist.groupby("customer_unique_id").agg(
    {
        "order_purchase_timestamp": lambda x: (pin_date - x.max()).days,
        "nb_orders": "nunique",
        "payment_value": "sum",
        "review_score": "mean",
        "payment_installments": "mean",
        "payment_sequential": "mean",
        "%fret": "mean",
        "day_diff_achat": "mean",
    }
)
# rename the columns
df_olist_2.rename(
    columns={
        "order_purchase_timestamp": "Recency",
        "nb_orders": "Frequency",
        "payment_value": "Monetary",
        "review_score": "Score_moyen",
        "payment_installments": "Nb_Versement",
        "payment_sequential": "Moyen_paiement",
        "%fret": "Fret_moyen",
    },
    inplace=True,
)
df_olist_2.head()
```

Pour déterminer le moment où les clients changent de cluster, nous allons itérer le K-Means sur toute la période avec des deltas de 7jours et calculer le score ARI

```python
# Pour trouver ma valeur de start j'ai mis en place la date la plus récente possible+270 soit 11 mois
start_date = df_olist_2["Recency"].min() + 270
# définir la première période

df_i = df_olist_2[df_olist_2["Recency"] <= start_date]
df_i.head(100)
```

Une fonction qui definie la période de référence sur laquelle je lance un kmeans et je prédis
une période de refrence+1 date ( 1 semaine), sur laquelle je lance un kmeans et je prédis
a chaque fois je scale 
Je comprAe le score ARI

```python
def maintenance(date):
    # date depart :
    start_date = df_olist_2["Recency"].min() + 270

    # définir la première période

    df_i = df_olist_2[df_olist_2["Recency"] <= start_date]
    print("Taille df de reference : ", df_i.shape[0])
    # Standarisation
    qt = QuantileTransformer(n_quantiles=900, random_state=42)
    scaler_init = qt.fit(df_i)
    df_i_s = scaler_init.transform(df_i)

    # transform into a dataframe
    df_i_s = pd.DataFrame(df_i_s, index=df_i.index, columns=df_i.columns)
    df_i_s.reset_index(inplace=True)

    # période incrementée
    df_m = df_olist_2[df_olist_2["Recency"] <= start_date + date]
    print("Taille nouveau df : ", df_m.shape[0])

    # standarisation
    df_m_s = qt.fit_transform(df_m)
    df_m_s = pd.DataFrame(df_m_s, index=df_m.index, columns=df_m.columns)
    df_m_s.reset_index(inplace=True)

    # meme index
    df_m_s = df_m_s[df_m_s.index.isin(df_i_s.index)]

    # kmeans initiale
    KMeans_init = KMeans(5, random_state=42)
    # Je fite sur le modèle initiale
    KMeans_init = KMeans_init.fit(df_i_s.iloc[:, 1:9])

    # Kmeans maintenance
    KMeans_aug = KMeans(5, random_state=42)
    # Je fite df+mois
    KMeans_aug = KMeans_aug.fit(df_m_s.iloc[:, 1:9])

    # Predictions
    df_m_s_init = scaler_init.transform(df_m)
    df_m_s_init = pd.DataFrame(df_m_s_init, index=df_m.index, columns=df_m.columns)
    df_m_s_init.reset_index(inplace=True)

    # meme index
    df_m_s_init = df_m_s_init[df_m_s_init.index.isin(df_i_s.index)]
    # predictions
    C_init = KMeans_init.predict(df_m_s_init.iloc[:, 1:9])
    C_pred = KMeans_aug.predict(df_m_s.iloc[:, 1:9])
    # calcul score ARI
    ari_p = round(metrics.adjusted_rand_score(C_pred, C_init), 2)
    print("ARI: ", ari_p)

    return ari_p
```

Faire tourner la fonction en boucle : 
Test 7jours après avec une cadence de 1 semaine 

```python
dates = np.arange(7, 330, 7)
ari_s = []

for date in dates:

    ari_p = maintenance(date)
    ari_s.append(ari_p)

plt.figure(figsize=(12, 8))
plt.plot(dates, ari_s)
plt.xlabel("Nombre de jours")
plt.ylabel("ARI")
plt.show()
```
