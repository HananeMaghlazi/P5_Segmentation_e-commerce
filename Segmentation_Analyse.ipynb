### Vérification de la distribution des données: <a class="anchor" id="dist"></a>

```python
col = ["price", "freight_value"]
# Distribution des variables
fig = plt.figure(1, figsize=(14, 8))
for i, label in enumerate(col):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df_order_items[label])
    plt.subplot(2, 2, i + 3)
    sns.boxplot(df_order_items[label])
plt.show()
```

- La plupart des commandes sont inférieures à 1000
- La plupart des valeurs de fret sont inférieures à 100
- Des commandes à prix et fret élevés, en nombre non négligeable, rendent la distribution asymétrique
- Beaucoup d'outliers detectés sur le prix et la valeur de fret
- J'utilise la transformation en log pour la visualisation des distributions

```python
fig = plt.figure(1, figsize=(14, 8))
for i, label in enumerate(col):
    plt.subplot(2, 2, i + 1)
    sns.histplot(
        df_order_items[label].apply(lambda x: np.log(1 + x)),
        bins=int(1 + np.log2(len(df_order_items))),
    )
    plt.subplot(2, 2, i + 3)
    sns.boxplot(df_order_items[label].apply(lambda x: np.log(1 + x)))
plt.show()
```

Avec la transformation log les variables présentent des distributions normales, on peut mieux voire également les outliers

### Identification et gestion des valeurs atypiques avec la règle 1,5 x écart interquartile : <a class="anchor" id="eir"></a>

```python
#  Appel de la fonction outliers
df_outlier = outliers(df_olist, ["price"])
df_outlier.head()
```

- La méthode des interquartiles a detecté 8440 produit en tant qu'un outlier
- Je vérifie de quel produit il s'agit

```python
# calculate RFM values
rfm = df_olist_cleaned.groupby("customer_unique_id").agg(
    {
        "order_purchase_timestamp": lambda x: (pin_date - x.max()).days,
        "nb_orders": "nunique",
        "payment_value": "sum",
    }
)
# rename the columns
rfm.rename(
    columns={
        "order_purchase_timestamp": "Recency",
        "nb_orders": "Frequency",
        "payment_value": "Monetary",
    },
    inplace=True,
)
rfm.head()
```
